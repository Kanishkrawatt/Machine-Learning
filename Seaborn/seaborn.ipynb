```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
sns.set_theme()
```

```python
from tkinter.ttk import Style


sns.relplot(data=tips,x="total_bill",y="tip",col="time",hue="smoker",style="smoker",size='size')
```

```python
# load the Iris
Iris = sns.load_dataset('iris')
Iris.head()
```

```python
# Scatter Plot
sns.scatterplot(x="sepal_length",y="sepal_width",hue="species",data=Iris)
```

```python
# Scatter Plot
sns.scatterplot(x="petal_length",y="petal_width",hue="species",data=Iris)
```

```python
# Loading the Titanic Dataset
Titanic = sns.load_dataset('titanic')
Titanic.head()
```

```python
# count Plot
sns.countplot(x='class',data=Titanic)
```

```python
sns.countplot(x='survived',data=Titanic)
```

```python
# BarPlot
sns.barplot(x='sex',y='survived',hue='class',data=Titanic)
```

```python
from sklearn.datasets import load_boston
house_boston = load_boston()

house = pd.DataFrame(house_boston.data, columns=house_boston.feature_names)
house['PRICE'] = house_boston.target
```

```python
house.head()
```

```python
sns.displot(house['PRICE'],kde=True)
```

```python
# Correlation Matrix
corr = house.corr()
corr
```

```python
# Heatmap

plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True,square=True,cmap='Blues',cbar=True,fmt='.2f',annot_kws={'size':10})
```
