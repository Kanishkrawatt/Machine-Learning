```python
# import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

Plotting the Data

```python
# Sine Wave
plt.plot(x,y)
plt.show()
```

```python
# Cosine Wave
plt.plot(x,z)
plt.show()
```

```python
# Adding Title x-axis and y-axis label
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sine Value")
plt.title("Sine Wave")
plt.show()
```

```python
# Parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"r*")
plt.show()
```

```python
plt.plot(x,y,"rx")
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),"g-")
plt.plot(x,np.cos(x),"r--")
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ["English","French","Spanish","Latin"]
people=[100,50,150,40]
ax.bar(languages,people)
plt.xlabel("Languages")
plt.ylabel("Number of People")
plt.show()
```

Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ["English","French","Spanish","Latin"]
people=[100,50,150,40]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color="g")
ax.scatter(x,z,color="r")
plt.show()
```

3d Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection="3d")
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
plt.show()
```
